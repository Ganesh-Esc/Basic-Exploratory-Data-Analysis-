```python
import pandas as pd
```

```python
df=pd.read_csv("Fact.FactPlanningMeasures.csv")
```

```python
df
```

```python
df.set_index('Time.[Month]',inplace=True)
```

```python
df
```

```python
df.groupby('Customer.[Account]').sum().head()
```

```python
df.index = pd.to_datetime(df.index)
df.groupby(['Customer.[Account]','Product.[SKU]']).resample('ME').sum()
```

```python
new_df=df[(df['Customer.[Account]']=='Walmart')& (df['Product.[SKU]']=='KitKat')]
```

```python
new_df.head()
```

```python
new_df['Cumulative Summed Budget']=new_df['Budget'].cumsum()
```

```python
out_df=new_df.reset_index()
```
